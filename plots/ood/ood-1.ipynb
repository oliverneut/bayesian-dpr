```python
import matplotlib.pyplot as plt
import numpy as np
from omegaconf import OmegaConf
args = OmegaConf.load('../colors.yml')
```

```python
models = ['DPR-T', 'DPR', 'DPR-VBLL-0.0', 'DPR-VBLL-0.001', 'DPR-VBLL-0.001 (KL)']
ndcg_nq = [0.4571, 0.4024, 0.0300, 0.1146, 0.0124]
mrr_nq = [0.4199, 0.3676, 0.0125, 0.0516, 0.0099]

ndcg_hotpotqa = [0.5782, 0.4856, 0.2830, 0.3636, 0.0240]
mrr_hotpotqa = [0.7770, 0.6774, 0.3407, 0.5052, 0.0273]

ndcg_fiqa = [0.3282, 0.2648, 0.2122, 0.1935, 0.0240]
mrr_fiqa = [0.3968, 0.3313, 0.2657, 0.2475, 0.0273]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 6.5), sharey=True)

bar_width = 0.35
x = np.arange(len(models))

# Colors
ndcg_color = args.red.mid
mrr_color = args.blue.mid
ndcg_border = args.red.dark
mrr_border = args.blue.dark

# Natural Questions
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].bar(x - bar_width/2, ndcg_nq, width=bar_width, color=ndcg_color, label='nDCG@10', edgecolor=ndcg_border)
axes[0].bar(x + bar_width/2, mrr_nq, width=bar_width, color=mrr_color, label='MRR@10', edgecolor=mrr_border)
axes[0].set_title('Natural Questions')
axes[0].set_xticks(x)
axes[0].set_xticklabels(models, rotation=20)
# axes[0].legend()
axes[0].grid(True, axis='y', ls='--')

# HotpotQA
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].bar(x - bar_width/2, ndcg_hotpotqa, width=bar_width, color=ndcg_color, label='nDCG@10', edgecolor=ndcg_border)
axes[1].bar(x + bar_width/2, mrr_hotpotqa, width=bar_width, color=mrr_color, label='MRR@10', edgecolor=mrr_border)
axes[1].set_title('HotpotQA')
axes[1].set_xticks(x)
axes[1].set_xticklabels(models, rotation=20)
# axes[1].legend()
axes[1].grid(True, axis='y', ls='--')

# FiQA2018
axes[2].spines['top'].set_visible(False)
axes[2].spines['right'].set_visible(False)
axes[2].bar(x - bar_width/2, ndcg_fiqa, width=bar_width, color=ndcg_color, label='nDCG@10', edgecolor=ndcg_border)
axes[2].bar(x + bar_width/2, mrr_fiqa, width=bar_width, color=mrr_color, label='MRR@10', edgecolor=mrr_border)
axes[2].set_title('FiQA2018')
axes[2].set_xticks(x)
axes[2].set_xticklabels(models, rotation=20)
axes[2].legend()
axes[2].grid(True, axis='y', ls='--')

axes[0].set_ylabel('Score')
plt.tight_layout()
plt.show()
```
