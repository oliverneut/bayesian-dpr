```python
import matplotlib.pyplot as plt
import numpy as np
from omegaconf import OmegaConf
args = OmegaConf.load('../colors.yml')
```

```python
models = ['DPR-T', 'DPR', 'DPR-VBLL-0.0', 'DPR-VBLL-0.001', 'DPR-VBLL-0.001 (KL)']
ndcg_nq = [0.4571, 0.4076, 0.4008, 0.3990, 0.2707]
mrr_nq = [0.4199, 0.3740, 0.3665, 0.3641, 0.2365]

ndcg_hotpotqa = [0.5782, 0.4974, 0.4802, 0.4714, 0.4392]
mrr_hotpotqa = [0.7770, 0.6940, 0.6712, 0.6651, 0.6144]

ndcg_fiqa = [0.3282, 0.2543, 0.2592, 0.2602, 0.2013]
mrr_fiqa = [0.3968, 0.3169, 0.3276, 0.3257, 0.2517]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 6.5), sharey=True)

bar_width = 0.3
x = np.arange(len(models))

# Colors
# ndcg_color = args.red.mid
# mrr_color = args.blue.mid
ndcg_color = args.red.mid
mrr_color = args.blue.mid
ndcg_border = args.red.dark
mrr_border = args.blue.dark

# Natural Questions
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].bar(x - bar_width/2, ndcg_nq, width=bar_width, color=ndcg_color, label='nDCG@10', edgecolor=ndcg_border, linewidth=1.2)
axes[0].bar(x + bar_width/2, mrr_nq, width=bar_width, color=mrr_color, label='MRR@10', edgecolor=mrr_border, linewidth=1.2)
axes[0].set_title('Natural Questions')
axes[0].set_xticks(x)
axes[0].set_xticklabels(models, rotation=20)
axes[0].set_ylim(0, 1)
axes[0].grid(True, axis='y', ls='--')

# HotpotQA
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].bar(x - bar_width/2, ndcg_hotpotqa, width=bar_width, color=ndcg_color, label='nDCG@10', edgecolor=ndcg_border, linewidth=1.2)
axes[1].bar(x + bar_width/2, mrr_hotpotqa, width=bar_width, color=mrr_color, label='MRR@10', edgecolor=mrr_border, linewidth=1.2)
axes[1].set_title('HotpotQA')
axes[1].set_xticks(x)
axes[1].set_xticklabels(models, rotation=20)
axes[1].set_ylim(0, 1)
axes[1].grid(True, axis='y', ls='--')

# FiQA2018
axes[2].spines['top'].set_visible(False)
axes[2].spines['right'].set_visible(False)
axes[2].bar(x - bar_width/2, ndcg_fiqa, width=bar_width, color=ndcg_color, label='nDCG@10', edgecolor=ndcg_border, linewidth=1.2)
axes[2].bar(x + bar_width/2, mrr_fiqa, width=bar_width, color=mrr_color, label='MRR@10', edgecolor=mrr_border, linewidth=1.2)
axes[2].set_title('FiQA2018')
axes[2].set_xticks(x)
axes[2].set_xticklabels(models, rotation=20)
axes[2].set_ylim(0, 1)
axes[2].legend()
axes[2].grid(True, axis='y', ls='--')

axes[0].set_ylabel('Score')
plt.tight_layout()
plt.show()
```
