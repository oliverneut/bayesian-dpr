```python
import matplotlib.pyplot as plt
import numpy as np
from omegaconf import OmegaConf
args = OmegaConf.load('../colors.yml')
```

```python
alpha = [0.0, 0.001, 0.01, 0.1, 1.0]
ndcg_msmarco = [0.330614779953426, 0.334252160091072, 0.100971673285095, 0.01395437362564, 0.332936097455837]
mrr_msmarco = [0.27618797471233, 0.280523491153863, 0.0842421089734843, 0.0119421703734024, 0.278202858507299]

ndcg_trec_dl_19 = [0.330614779953426, 0.334252160091072, 0.100971673285095, 0.01395437362564, 0.332936097455837]
mrr_trec_dl_19 = [0.27618797471233, 0.280523491153863, 0.0842421089734843, 0.0119421703734024, 0.278202858507299]

ndcg_trec_dl_20 = [0.330614779953426, 0.334252160091072, 0.100971673285095, 0.01395437362564, 0.332936097455837]
mrr_trec_dl_20 = [0.27618797471233, 0.280523491153863, 0.0842421089734843, 0.0119421703734024, 0.278202858507299]
```

```python
ndcg_dark = args.red.dark
ndcg_mid = args.red.mid

mrr_dark = args.blue.dark
mrr_mid = args.blue.mid

fig, ax = plt.subplots(1, 3, figsize=(18, 6))
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)

positions = range(len(alpha))  # [0, 1, 2, 3, 4]
ax[0].plot(positions, ndcg_msmarco, marker='o', color=args.red.dark, markerfacecolor=ndcg_mid, markeredgecolor=ndcg_dark, label='nDCG@10')
ax[0].plot(positions, mrr_msmarco, marker='s', color=args.blue.dark, markerfacecolor=mrr_mid, markeredgecolor=mrr_dark, label='MRR@10')
ax[0].set_xlabel('$\\alpha$', fontsize=14)
ax[0].set_ylabel('Performance', fontsize=14)
ax[0].set_title('MS MARCO', fontsize=16)
ax[0].set_xticks(positions, alpha)
ax[0].legend()
ax[0].grid(True, which="both", ls="--")

ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)

ax[1].plot(positions, ndcg_trec_dl_19, marker='o', color=args.red.dark, markerfacecolor=ndcg_mid, markeredgecolor=ndcg_dark, label='nDCG@10')
ax[1].plot(positions, mrr_trec_dl_19, marker='s', color=args.blue.dark, markerfacecolor=mrr_mid, markeredgecolor=mrr_dark, label='MRR@10')
ax[1].set_xlabel('$\\alpha$', fontsize=14)
ax[1].set_title('TREC DL 2019', fontsize=16)
ax[1].set_xticks(positions, alpha)
ax[1].legend()
ax[1].grid(True, which="both", ls="--")


ax[2].spines['top'].set_visible(False)
ax[2].spines['right'].set_visible(False)

ax[2].plot(positions, ndcg_trec_dl_20, marker='o', color=args.red.dark, markerfacecolor=ndcg_mid, markeredgecolor=ndcg_dark, label='nDCG@10')
ax[2].plot(positions, mrr_trec_dl_20, marker='s', color=args.blue.dark, markerfacecolor=mrr_mid, markeredgecolor=mrr_dark, label='MRR@10')
ax[2].set_xlabel('$\\alpha$', fontsize=14)
ax[2].set_title('TREC DL 2020', fontsize=16)
ax[2].set_xticks(positions, alpha)
ax[2].legend()
ax[2].grid(True, which="both", ls="--")
```
